```python
# Ejercicio 1: mostrar las primeras filas del DataFrame
print('Ejercicio 1: mostrar las primeras filas del DataFrame')
print(df.head(18))
#print("/n")
```

```python
# Ejercicio 2: calcular el total de ventas por regiòn
print("Ejercicio 2: calcular el total de ventas por regiòn")
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print("/n")
```

```python
# Ejercicio 3: filtar el DataFrame para mostrar solo las ventas realizadas por David
print("Ejercicio 3: filtar el DataFrame para mostrar solo las ventas realizadas por David")
ventas_david = df[df['Vendedor'] == 'David']
print(ventas_david)
print("/n")
```

```python
# Ejercicio 4: calcular el promedio de ventas por producto
print("Ejercicio 4: calcular el promedio de ventas por producto")
ventas_por_producto = df.groupby ('Producto')['Ventas'].mean()
print(ventas_por_producto)
print("/n")
```

```python
# Ejercicio 5: ordenar el DataFrame por ventas en orden descendente
print("Ejercicio 5: ordenar el DataFrame por ventas en orden descendente ")
df_ordenado = df.sort_values(by='Ventas',ascending=False)
print(df_ordenado)
print("/n")
```

**Manipulaciòn de Datos**

*   Filtrar las filas donde las ventas sean mayores a 10000
*   Ordena el DataFrame por la columna "Ventas" en orden descendente
*   Elimina la columna "Vendedor"




```python
# Ejericio 6: Filtrar las filas con ventas mayores a 10000
print("Ejericio 6: Filtrar las filas con ventas mayores a 10000")
df_filtrado = df[df['Ventas'] > 10000]
print(df_filtrado)
print("/n")
```

```python
# Ejercicio 7: Ordenar el DataFrame por ventas en orden descendente
print("Ejercicio 7: Ordenar el DataFrame por ventas en orden descendente")
df_ordenado = df.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
print("/n")
```

```python
# Ejercicio 7.1: Ordenar el DataFrame por ventas en orden ascendente
print("Ejercicio 7.1: Ordenar el DataFrame por ventas en orden ascendente")
df_ordenado = df.sort_values(by='Ventas', ascending=True)
print(df_ordenado)
print("/n")
```

```python
# Ejercicio 8: Eliminar columna "Vendedor"
print("Ejercicio 8: Eliminar columna 'Vendedor'")
df_sin_vendedor = df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
print("/n")
```

**Agregaciòn y Agrupaciòn de Datos**

*   Calcular el total de ventaspor regiòn

```python
# Ejercicio 9: Calcular el valor total de ventas por regiòn
print("Ejercicio 9: Calcular el valor total de ventas por regiòn ")
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print("ventas_por_region")
print("/n")
```

```python
# Ejercicio 10: Calcular el total de ventas por producto
print("Ejercicio 10: Calcular el total de ventas por producto")
ventas_por_producto = df.groupby('Producto')['Ventas'].sum()
print("ventas_por_producto")
print("/n")
```
